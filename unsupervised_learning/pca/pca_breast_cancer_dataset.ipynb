```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

print("Dataset Shape:", X.shape)
print("Target Classes:", data.target_names)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
model_no_pca = LogisticRegression(max_iter=5000)
model_no_pca.fit(X_train, y_train)

y_pred_no_pca = model_no_pca.predict(X_test)
acc_no_pca = accuracy_score(y_test, y_pred_no_pca)
print("Accuracy WITHOUT PCA:", acc_no_pca)
```

```python
pca_2 = PCA(n_components=2)
X_train_pca2 = pca_2.fit_transform(X_train)
X_test_pca2 = pca_2.transform(X_test)

model_pca2 = LogisticRegression(max_iter=5000)
model_pca2.fit(X_train_pca2, y_train)

y_pred_pca2 = model_pca2.predict(X_test_pca2)
acc_pca2 = accuracy_score(y_test, y_pred_pca2)

print("Accuracy WITH PCA (2 Components):", acc_pca2)
```

```python
pca_3 = PCA(n_components=3)
X_train_pca3 = pca_3.fit_transform(X_train)
X_test_pca3 = pca_3.transform(X_test)

model_pca3 = LogisticRegression(max_iter=5000)
model_pca3.fit(X_train_pca3, y_train)

y_pred_pca3 = model_pca3.predict(X_test_pca3)
acc_pca3 = accuracy_score(y_test, y_pred_pca3)

print("Accuracy WITH PCA (3 Components):", acc_pca3)
```

```python
pca_4 = PCA(n_components=4)
X_train_pca4 = pca_4.fit_transform(X_train)
X_test_pca4 = pca_4.transform(X_test)

model_pca4 = LogisticRegression(max_iter=5000)
model_pca4.fit(X_train_pca4, y_train)

y_pred_pca4 = model_pca4.predict(X_test_pca4)
acc_pca4 = accuracy_score(y_test, y_pred_pca4)

print("Accuracy WITH PCA (4 Components):", acc_pca4)
```

```python
pca = PCA(n_components=13)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

model_pca = LogisticRegression(max_iter=5000)
model_pca.fit(X_train_pca, y_train)

y_pred_pca = model_pca.predict(X_test_pca)
acc_pca = accuracy_score(y_test, y_pred_pca)

print("Accuracy WITH PCA:", acc_pca)
```

```python
print("\nExplained Variance (2 components):", pca_2.explained_variance_ratio_)
print("Total Variance (2 components):", sum(pca_2.explained_variance_ratio_))

print("\nExplained Variance (3 components):", pca_3.explained_variance_ratio_)
print("Total Variance (3 components):", sum(pca_3.explained_variance_ratio_))

print("\nExplained Variance (4 components):", pca_4.explained_variance_ratio_)
print("Total Variance (4 components):", sum(pca_4.explained_variance_ratio_))
```

```python
def manual_pca(X, n_components):
    mean = np.mean(X, axis=0)
    X_centered = X - mean
    covariance_matrix = np.cov(X_centered.T)
    eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)
    
    sorted_index = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_index]
    sorted_eigenvectors = eigenvectors[:, sorted_index]
    
    eigenvector_subset = sorted_eigenvectors[:, 0:n_components]
    
    X_reduced = np.dot(X_centered, eigenvector_subset)
    
    return X_reduced, sorted_eigenvalues, eigenvector_subset, mean
```

```python
X_train_pca3, eigenvalues, eigenvectors, mean = manual_pca(X_train, 3)

X_test_centered = X_test - mean
X_test_pca3 = np.dot(X_test_centered, eigenvectors)

model_pca3 = LogisticRegression(max_iter=5000)
model_pca3.fit(X_train_pca3, y_train)

y_pred_pca3 = model_pca3.predict(X_test_pca3)

print("Accuracy WITH PCA (3 Components):",
      accuracy_score(y_test, y_pred_pca3))
```

```python
X_train_pca3, eigenvalues, eigenvectors, mean = manual_pca(X_train, 4)

X_test_centered = X_test - mean
X_test_pca3 = np.dot(X_test_centered, eigenvectors)

model_pca3 = LogisticRegression(max_iter=5000)
model_pca3.fit(X_train_pca3, y_train)

y_pred_pca3 = model_pca3.predict(X_test_pca3)

print("Accuracy WITH PCA (4 Components):",
      accuracy_score(y_test, y_pred_pca3))
```

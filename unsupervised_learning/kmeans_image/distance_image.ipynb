```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#load the static images
img1=cv2.imread('image_1.png')
img1_rgb=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1_rgb)
plt.axis('off')
plt.title('Image 1')
plt.show()
```

```python
img2=cv2.imread('image_2.png')
img2_rgb=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2_rgb)
plt.axis('off')
plt.title('Image 2')
plt.show()
```

```python
img1_rgb.shape
```

```python
img2_rgb.shape
```

```python
#resizing the images to a common size
img1_resized=cv2.resize(img1_rgb, (200,200))
img2_resized=cv2.resize(img2_rgb, (200,200))
print(f"resized image 1 shape: {img1_resized.shape}")
print(f"resized image 2 shape: {img2_resized.shape}")
```

```python
#flatten the image into vectors
img1_vector=img1_resized.flatten()
img2_vector=img2_resized.flatten()
print(f"image 1 vector shape: {img1_vector.shape}")
print(f"image 2 vector shape: {img2_vector.shape}")
```

```python
distance=np.linalg.norm(img1_vector-img2_vector)
print(f"Euclidean distance between the two images: {distance}")
```

```python
distance = np.linalg.norm(img1_vector - img2_vector, ord=1)
print(f"Manhattan distance between the two images: {distance}")
```

```python
p = 3  # you can change this value

distance = np.linalg.norm(img1_vector - img2_vector, ord=p)
print(f"Minkowski distance (p={p}) between the two images: {distance}")
```

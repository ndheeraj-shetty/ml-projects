```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
```

```python
df=load_breast_cancer()
x=df.data
y=df.target
```

```python
x.shape
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
u,s,vt=np.linalg.svd(x_scaled, full_matrices=False)
```

```python
print("shape of actual matrix:", x_scaled.shape)
print("shape of U matrix:", u.shape)
print("shape of s matrix:", s.shape)
print("shape of v transpose matrix:", vt.shape)
```

```python
k=2
x_reduced=u[:, :k]*s[:k]
print("reduced x shape: ", x_reduced.shape)
variance_explained=(s**2)*np.sum(s**2)
print("explained variance: ", variance_explained)
print("total variance: ", np.sum(variance_explained))
```

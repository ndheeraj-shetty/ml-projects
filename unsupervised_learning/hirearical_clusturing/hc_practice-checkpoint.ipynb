```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
```

```python
data=load_breast_cancer()
df=pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
scaler=StandardScaler()
df_scaled=scaler.fit_transform(df)
```

```python
z=linkage(df_scaled, method='ward')
```

```python
plt.figure(figsize=(10,6))
dendrogram(z, no_labels=True, color_threshold=12)
plt.axhline(y=12, color='red', linestyle='--', label='cut line(k=4)')
plt.ylabel("distance")
plt.title('dendrogram')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
model=AgglomerativeClustering(n_clusters=4, linkage='ward')
df['cluster']=model.fit_predict(df_scaled)
```

```python
print(df['cluster'].value_counts())
```

```python
print("silhauette score", silhouette_score(df_scaled, df['cluster']))
```

```python
colors=['red', 'green', 'blue', 'yellow']
for c in range(4):
    subset=df[df['cluster']==c]
    plt.scatter(subset['fractal dimension error'],subset['worst compactness'], color=colors[c],alpha=0.6)
plt.xlabel('fractal dimension error')
plt.ylabel('worst compactness')
plt.title('agglomerative ckustering')
plt.legend()
plt.show()
```
